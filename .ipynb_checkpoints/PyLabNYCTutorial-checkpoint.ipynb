```python
#value_counts() counts the frequencies of the variables and sorts them in desceding order
 
popular_zip_codes = orig_data['Incident Zip'].value_counts()[:10]
popular_zip_codes
#".index" returns the index
popular_zip_codes = orig_data['Incident Zip'].value_counts()[:10].index
popular_zip_codes

```

```python
#groupby builds a stacked frequency table with the first column as the first key and second column the second key
zipcode_incident_table = orig_data.groupby(['Incident Zip', 'Complaint Type'])['Descriptor'].aggregate(len)
zipcode_incident_table[:10]
```

```python
#unstack() builds a 2D table
zipcode_incident_table = orig_data.groupby(['Incident Zip', 'Complaint Type'])['Descriptor'].aggregate(len).unstack()
zipcode_incident_table[:10]
```

```python
top_5_complaints = zipcode_incident_table.transpose()[popular_zip_codes]
top_5_complaints[:10]
```

```python
normalized_complaints = top_5_complaints / top_5_complaints.sum()
normalized_complaints[:10]
```

```python
#Need to drop NA and sort the columns to get the 5 most frequent complaint types - we sort along a single index
#we then transpose it to get the xip codes per rows
x = normalized_complaints.dropna(how='any').sort('11226', ascending=False)[:5].transpose()
x[:10]
```

```python
normalized_complaints.dropna(how='any').sort('11226', ascending=False)[:5].transpose().plot(kind='bar')
```

```python
complaints = orig_data[['Created Date', 'Complaint Type']]
noise_complaints = complaints[complaints['Complaint Type'] == 'Noise - Street/Sidewalk']
#By deafult index is on the record type, we need to set it on the date of creation to do tiime series anlysis
# we first set_index, the we sort along that index, then we resample it along every hour, Can also use '5Min' '2Min' 'D'...
# This is a way of doing averaging
noise_complaints.set_index('Created Date').sort_index().resample('H', how=len).plot()
```

```python
complaints = orig_data[['Created Date', 'Complaint Type']]
noise_complaints = complaints[complaints['Complaint Type'] == 'Noise - Street/Sidewalk']
noise_complaints = noise_complaints.set_index("Created Date")
```

```python
#Create an index on date column and then we can access the weekday, hour through fields like index.weekday and index.hour
noise_complaints['weekday'] = noise_complaints.index.weekday
noise_complaints[:3]
```

```python
# Count the complaints by weekday
counts_by_weekday = noise_complaints.groupby('weekday').aggregate(len)
counts_by_weekday
```

```python
# change the index to be actual days
counts_by_weekday.index = ["Sunday", "Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday"]
```

```python
counts_by_weekday.plot(kind='bar')
```

```python
complaints = orig_data[['Created Date', 'Complaint Type']]
noise_complaints = complaints[complaints['Complaint Type'] == 'Noise - Street/Sidewalk']
noise_complaints = noise_complaints.set_index("Created Date")
# Count the complaints by hour
noise_complaints['hour'] = noise_complaints.index.hour
noise_complaints[:3]
counts_by_hour = noise_complaints.groupby('hour').aggregate(len)
counts_by_hour
counts_by_hour.plot(kind='bar')
```

```python
# We need to get rid of the NA values for this to work
street_names = orig_data['Street Name'].fillna('')

manhattan_streets = street_names[street_names.str.contains("MANHATTAN")]
manhattan_streets
```

```python
# Our current latitude and longitude
our_lat, our_long = 40.714151,-74.00878
distance_from_us = (orig_data['Longitude'] - our_long)**2 + (orig_data['Latitude'] - our_lat)**2
#In case the variable is continous we need to plot a histogram and for that we need to convert it into a Series object
pd.Series(distance_from_us).hist()
```

```python
close_complaints = orig_data[distance_from_us < 0.00005]
close_complaints['Complaint Type'].value_counts()[:20].plot(kind='bar')
```
